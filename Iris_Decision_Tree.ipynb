```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
```

```python
X = iris.data[:, 2:] # Petal lenght and width
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X,y)
```

```python
tree_clf.predict_proba([[5,1.5]])
```

```python
tree_clf.predict([[5,1.5]])
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
X = x_base.reshape(-1,1)
y = y_base
```

```python
def forest_regressor(max_depth,min_samples_leaf=1):
    tree_reg = DecisionTreeRegressor(max_depth=max_depth,min_samples_leaf=min_samples_leaf)
    tree_reg.fit(X,y)
    y_pred = tree_reg.predict(X)
    
    plt.scatter(x_base,y_base)
    plt.plot(x_base,y_pred,color = 'red')
```

```python
forest_regressor(max_depth=2)
```
